#Computer Vision - CNN


The goal of this task is to get hands-on experience in developing, training and testing Convolutional Neural Network for the Computer Vision task of Object/Image Classification. Uses the CIFAR-100 dataset.

```python
#imports:
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
from google.colab import drive
from random import sample
drive.mount('/content/drive')
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.datasets import reuters
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
#Sources: Load Cifar100 data: https://www.tensorflow.org/api_docs/python/tf/keras/datasets/cifar100/load_data
# Load the data - training as well as testing
(ImageTrain, LabelTrain), (ImageTest, LabelTest) = tf.keras.datasets.cifar100.load_data()
ImageTrain, ImageTest = ImageTrain / 255.0, ImageTest / 255.0
# The data from TensforFlow and Keras will only have integer class labels. Each of those 100 integer class labels correspond to the following names, in the correct order
fine_labels = ['apple', 'aquarium_fish', 'baby', 'bear', 'beaver', 'bed', 'bee', 'beetle', 'bicycle', 'bottle', 'bowl', 'boy', 'bridge', 'bus', 'butterfly', 'camel', 'can', 'castle', 'caterpillar', 'cattle', 'chair', 'chimpanzee', 'clock', 'cloud', 'cockroach', 'couch', 'crab', 'crocodile', 'cup', 'dinosaur', 'dolphin', 'elephant', 'flatfish', 'forest', 'fox', 'girl', 'hamster', 'house', 'kangaroo', 'keyboard', 'lamp', 'lawn_mower', 'leopard', 'lion', 'lizard', 'lobster', 'man', 'maple_tree', 'motorcycle', 'mountain', 'mouse', 'mushroom', 'oak_tree', 'orange', 'orchid', 'otter', 'palm_tree', 'pear', 'pickup_truck', 'pine_tree', 'plain', 'plate', 'poppy', 'porcupine', 'possum', 'rabbit', 'raccoon', 'ray', 'road', 'rocket', 'rose', 'sea', 'seal', 'shark', 'shrew', 'skunk', 'skyscraper', 'snail', 'snake', 'spider', 'squirrel', 'streetcar', 'sunflower', 'sweet_pepper', 'table', 'tank', 'telephone', 'television', 'tiger', 'tractor', 'train', 'trout', 'tulip', 'turtle', 'wardrobe', 'whale', 'willow_tree', 'wolf', 'woman', 'worm']

# These are the string labels for the 20 superclasses. You may not need to use this at all, just provided here for reference.
coarse_labels = ['aquatic_mammals', 'fish', 'flowers', 'food_containers', 'fruit_and_vegetables', 'household_electrical_devices', 'household_furniture', 'insects', 'large_carnivores', 'large_man-made_outdoor_things', 'large_natural_outdoor_scenes', 'large_omnivores_and_herbivores', 'medium_mammals', 'non-insect_invertebrates', 'people', 'reptiles', 'small_mammals', 'trees', 'vehicles_1', 'vehicles_2']

#Sources: How to plot imageshttps://matplotlib.org/stable/tutorials/images.html
# Visualize the data by plotting 100 random images, one each for the 100 classes
# Draw 10 images in one row, 10 rows total
ImagePos = []
for i in range(100):
    ImageID= np.where(LabelTrain == i)[0]
    ImagePos.append(sample(list(ImageID), 1)[0])
plt.figure(figsize=(15, 15))
for i, ImageID in enumerate(ImagePos):
    plt.subplot(10, 10, i+1) # 10x10
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(ImageTrain[ImageID], cmap=plt.cm.binary)
    plt.xlabel(fine_labels[LabelTrain[ImageID][0]])
plt.show()
```

# CNN Model Creation

 This will create the model used for the CNN.

```python
#Sources: model class, and summary class, also sequential keras: https://keras.io/api/models/model/
# Create a DL model for Computer Vision - Convolutional Neural Network (Use *TensorFlow* and *keras*, as shown in the example code in the lecture for 'deep-learning')
model = Sequential([Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)), MaxPooling2D(2, 2), Conv2D(64, (3, 3), activation='relu'),
                    MaxPooling2D(2, 2), Conv2D(64, (3, 3), activation='relu'), Flatten(), Dense(64, activation='relu'), Dense(100, activation='softmax')])

model.summary()
```

# Training

This section will handle training the model. Warning takes a while to train. Accuaacy goal is around 35%

```python
#Sources: model checkpoint tensorlfow: https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/ModelCheckpoint
# Set the model checkpoints to be saved in a folder in the google drive at at location "/content/drive/My Drive/Colab Notebooks/cifar/"
FilePath = "/content/drive/My Drive/Colab Notebooks/cifar/CifarModel.h5"
ModelWeights = ModelCheckpoint(filepath=FilePath, save_weights_only=True, verbose=1)

#sources: Adam optimizer: https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/Adam. fit method: https://keras.io/api/models/model_training_apis/
#it seems 'Adam' would be the best for this. It uses stochastic gradient decent and that is the method i know best compared to the others.
# Train/fit the DL model using the training CIFAR-100 data
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
CifarModelHist = model.fit(ImageTrain, LabelTrain, epochs=10, validation_data=(ImageTest, LabelTest), callbacks=[ModelWeights])

# Plot the training/validation accuracy and loss
#accuracy
plt.plot(CifarModelHist.history['val_accuracy'], label='Validation Accuracy', color='pink')
plt.plot(CifarModelHist.history['accuracy'], label='Training Accuracy', color='green')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')

#training andloss
plt.figure()
plt.plot(CifarModelHist.history['val_loss'], label='Validation Loss', color='pink')
plt.plot(CifarModelHist.history['loss'], label='Training Loss', color='green')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

# CNN Evaluation

This section will show the complete evaluation of the CNN model

```python
#sources: Sequential methodhttps://www.tensorflow.org/api_docs/python/tf/keras/Sequential
# Re-initialize the model
model = Sequential([Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)), MaxPooling2D(2, 2), Conv2D(64, (3, 3), activation='relu'),
                    MaxPooling2D(2, 2), Conv2D(64, (3, 3), activation='relu'), Flatten(), Dense(64, activation='relu'), Dense(100, activation='softmax')])

#found load weigths here: https://www.tensorflow.org/tutorials/keras/save_and_load, use .h5 for files for models: https://www.tensorflow.org/guide/keras/serialization_and_saving
# Load the model weights that were saved at - "/content/drive/My Drive/Colab Notebooks/cifar/"
SavedFilePath = "/content/drive/My Drive/Colab Notebooks/cifar/CifarModel.h5"
model.load_weights(SavedFilePath)

#optimize again but then print out the accuarcy
# Evaluate the trained DL model on the CIFAR-100 test dataset
#need this for loss value: https://www.tensorflow.org/api_docs/python/tf/keras/metrics/sparse_categorical_crossentropy
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
CifarLoss, CifarAcc = model.evaluate(ImageTest, LabelTest)
print(f"Accuracy: {CifarAcc * 100}%")
```

# Natural Language Processing - RNN

This section will cover The goal of this task is to get hands-on experience in developing, training and testing Recurrent Neural Network for the Natural Language Processing task of Sentiment Analysis / Text Classification.

USES: Reuters newswire classification dataset.

# RNN Model Creation

This sections covers the RNN Model Creation

```python
# Load the Reuters dataset - use the Keras version
#   Select the vocabulary size while loading the data
#   The data will be loaded as integer representations for each word
#found the data set here: https://keras.io/api/datasets/reuters/, tensro flow recommends 10,000 most common words: https://www.tensorflow.org/api_docs/python/tf/keras/datasets/reuters/load_data
FilePath2 = "/content/drive/My Drive/Colab Notebooks/reuters/ReutersModel.h5"#put ehre as a global for ease of use in testing adn training
VocabSize = 10000
(DataTrained, LabelTrained), (TestData, TestLabel) = reuters.load_data(num_words=VocabSize) #need to lload the reuter data.

# Prepare the data to be used for the next steps
#   Each data entry (newswire) can be of different lengths
#   Make each newswire consistent - same number of words
#   'Pad' words (say 0) to get to the standard length or remove words
# used this for padding: https://www.tensorflow.org/api_docs/python/tf/keras/utils/pad_sequences
NewsWireLen = 200#stadnard lenfth
DataTrained = pad_sequences(DataTrained, maxlen=NewsWireLen)
TestData = pad_sequences(TestData, maxlen=NewsWireLen)

# Create a DL model for Natural Language Processing - Recurrent Neural Network / LSTM / GRU
#Sources: LSTM units: https://stackoverflow.com/questions/59072728/what-is-the-rule-to-know-how-many-lstm-cells-and-how-many-units-in-each-lstm-cel, Embedding layer: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding
EmbedSize = 64  #embedding size
lstm = 128    #LSTM units
Model2 = Sequential([Embedding(VocabSize, EmbedSize, input_length=200), LSTM(lstm), Dense(46, activation='softmax')])

# Print the DL model summary
Model2.summary()
```

# Training

This Section handles the Training of the RNN model.

```python
# Set the model checkpoints to be saved in a folder in the google drive at at location "/content/drive/My Drive/Colab Notebooks/reuters/"
ModelWeights2 = ModelCheckpoint(filepath=FilePath2, save_weights_only=True, verbose=1)#saves modelweigths for filepath2

Model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
ReuterModelHist = Model2.fit(DataTrained, LabelTrained, epochs=5, validation_data=(TestData, TestLabel), callbacks=[ModelWeights2])

# Plot the training/validation accuracy and loss
#accuracy
plt.plot(ReuterModelHist.history['accuracy'], label='Training Accuracy', color='pink')
plt.plot(ReuterModelHist.history['val_accuracy'], label='Validation Accuracy', color='green')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

#training andloss
plt.figure()
plt.plot(ReuterModelHist.history['loss'], label='Training Loss', color='pink')
plt.plot(ReuterModelHist.history['val_loss'], label='Validation Loss', color='green')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Evaluation

This Section will shwo the overall Evaluation of the RNN model

```python
# Re-initialize the model
StandLen = 200
Model2 = Sequential([Embedding(VocabSize, EmbedSize, input_length=StandLen), LSTM(lstm), Dense(46, activation='softmax')])

# Load the model weights that were saved at - "/content/drive/My Drive/Colab Notebooks/reuters/"
Model2.load_weights(FilePath2)

#same thing as before but for nmodel2(Reuter) this time.
# Evaluate the trained DL model on the Reuters test dataset
Model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
ReuterLoss, ReuterAcc = Model2.evaluate(TestData, TestLabel)
print(f"Test Accuracy: {ReuterAcc * 100}%")
```
